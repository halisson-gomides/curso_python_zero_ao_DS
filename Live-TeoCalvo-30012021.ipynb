#### Quanto cada partido teve de votos totais em cada ano de eleicao em cada turno?

```python
df_group = df_votacao.groupby(['ano_eleicao', 'numero_turno', 'sigla_partido'])['total_votos'].sum()
print(df_group.head())
print(df_group.loc[2014][1]['PT'])
```

```python
df_group = df_group.reset_index()
df_group.sort_values ( by = ["ano_eleicao", "total_votos"], ascending = [True, False])
```

```python
df_group = df_group.sort_values ( by = ["ano_eleicao", "total_votos"], ascending = [True, False])
df_group = df_group.set_index ([ "ano_eleicao", "numero_turno", "sigla_partido"])
df_group
```

```python
# Group by com transposse

df_unstack = df_group.unstack().fillna(0)
df_unstack
```

```python
df_unstack.stack()
```

```python
# Calcular a representatividade de cada político no total de votos do seu respectivo partido político

df1 = df_group.reset_index()
df_votos = df_votacao.merge(right=df1, on=['ano_eleicao', 'numero_turno', 'sigla_partido'], how='left',
                      suffixes=['','_partido'])
df_votos['representacao_partido'] = df_votos['total_votos'] / df_votos['total_votos_partido']
df_votos
```

```python
import sqlalchemy as sql
```

```python
# criando a engine de conexão
con = sql.create_engine('sqlite:///dados/originais/dados-teo/banco.db')

# Tabelas que estão presentes no banco
con.table_names()
```

```python
df_bens  = pd.read_sql_table('tb_bens', con)
df_votos = pd.read_sql_table('tb_votacao', con)
```

```python
df_head = pd.read_sql_query('SELECT * FROM tb_votacao limit 10', con)
df_head
```

```python
query = '''
select 
    t1.ano_eleicao,
    t1.sigla_partido,
    t1.nome,
    t1.numero_sequencial,
    t2.valor
from tb_votacao as t1
left join tb_bens as t2
on t1.ano_eleicao = t2.ano_eleicao
and t1.descricao_eleicao  = t2.descricao_eleicao
and t1.numero_sequencial  = t2.numero_sequencial

where t1.numero_turno = 1

group by t1.ano_eleicao,
        t1.sigla_partido,
        t1.nome,
        t1.numero_sequencial
'''
df_resposta = pd.read_sql_query(query, con)
```

```python
df_analisys = df_resposta.groupby(['ano_eleicao', 'sigla_partido'])['valor'].sum().reset_index()
df_analisys
```

```python
df_analisys.to_sql('tb_patrimonio_partido', con, if_exists='replace', index=False)
```
